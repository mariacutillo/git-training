### Generiamo una matrice di numeri casuali e associamo una label
- generiamo due gaussiane in dim = 2
- creiamo il vettore di label
- facciamo un plot
- salviamo il dataset come dizionario

```python
np.ones((4,4)) #matrice 4 per 4 di tutti 1
```

```python
Npoints= 1000 #punti
D = 2 #dimensioni

mu1 = np.array([3,2])
mu2  =np.array([1,-1])
cov1 = np.array([[3, 1], [1, 2]])

d1 =  np.random.randn(Npoints, D)@ np.linalg.cholesky(cov1) + mu1  #a tutte le righe devo sommare lo stesso vettore
#stessa sintassi come np.ones((4,4))
d2 = 2*np.random.randn(Npoints, D) + mu2
```

```python
print(d1.shape)
```

```python
print(d1[2:100, :].shape) #slicing: prendiamo un array e lo facciamo a fette
d1.mean(axis=0)
```

```python
# cambiamo pacchetto e utilizziamo plt
f,ax = plt.subplots(1) #definisce gli assi
ax.scatter(d1[:,0],d1[:,1],s=5,alpha=0.3,color="r",label="d1") # s=quanto grandi i punti, alhpa=trasparenza
ax.scatter(d1[:,0].mean(),d1[:,1].mean(),s=100,alpha=1,marker="+",color="k",label="media d1") # marker + simbolo
ax.scatter(d2[:,0],d2[:,1],s=5,alpha=0.3,color="b",label="d2")
ax.scatter(d2[:,0].mean(),d2[:,1].mean(),s=100,alpha=1,marker="*",color="k",label="media d2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.axis('equal')
plt.legend()
```

```python
Npoints= 1000 #punti
D = 2 #dimensioni

mu1 = np.array([3,8])
mu2  =np.array([1,-1])
cov1 = np.array([[3, 1], [1, 2]])

d1 =  np.random.randn(Npoints, D)@ np.linalg.cholesky(cov1) + mu1  #a tutte le righe devo sommare lo stesso vettore
#stessa sintassi come np.ones((4,4))
d2 = 2*np.random.randn(Npoints, D) + mu2
```

```python
# cambiamo pacchetto e utilizziamo plt
f,ax = plt.subplots(1) #definisce gli assi
ax.scatter(d1[:,0],d1[:,1],s=5,alpha=0.3,color="r",label="d1") # s=quanto grandi i punti, alhpa=trasparenza
ax.scatter(d1[:,0].mean(),d1[:,1].mean(),s=100,alpha=1,marker="+",color="k",label="media d1") # marker + simbolo
ax.scatter(d2[:,0],d2[:,1],s=5,alpha=0.3,color="b",label="d2")
ax.scatter(d2[:,0].mean(),d2[:,1].mean(),s=100,alpha=1,marker="*",color="k",label="media d2")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.axis('equal')
plt.legend() #ho aumentato la distanza tra le medie quindi sono più facilmente separabili
```

```python
data= np.concatenate([d1,d2],axis=0)
labels = np.concatenate([np.zeros(Npoints),np.ones(Npoints)],axis=0)
print(data.shape)
print(labels.shape)
```

```python
np.savez("mydata.npz",data=data,labels=labels) # salva zippato nel folder
# np non può salvarlo direttamente in cvs
```

```python
from sklearn.model_selection import train_test_split # serve per tirare fuori tutti i modelli di ML

X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.33, random_state=42) # split casuale tra train e test
# il 33% lo mette nel test
# splitta allo stesso modo data e label
# se avessi avuto un terzo mettore avrei potuto aggiungere il terzo elemento dopo label e mi avrebbe fatto lo split anche su quello
# gli array devono essere tutti della stessa dimensione
#random_state cos'è?
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0) #definisce il modello clf, modello con 4 parametri
# i parametri prima del fit sono inizializzati casualmente
#print("pre fit", clf.coef_) # non posso farlo perchè non ho ancora creato i coefficienti
clf.fit(X_train, y_train) # in totale ha 2 parametri per la parte lineare (retta e pendenza) e 2 per il rumore
print("post fit", clf.coef_)
```

```python
clf.score(X_test, y_test)
```

```python
clf.coef_
```

```python
a1,a2 = clf.coef_[0]
b = clf.intercept_
m=-a1/a2
c=-b/a2
t=np.linspace(-4,6,100)
```

```python
pred_class = clf.predict(X_test)
```

```python
# problema di alto bias e poca varianza
pred_class = np.zeros(y_test.shape)
f,ax = plt.subplots(1)
n_fold = 20
for i  in range(n_fold-1):
  start = int(i*(Npoints/n_fold))
  stop = int((i+1)*(Npoints/n_fold))
  X_fold = X_train[start:stop,:]
  y_fold = y_train[start:stop]
  clf = LogisticRegression(random_state=0).fit(X_fold, y_fold)
  a1,a2 = clf.coef_[0]
  b = clf.intercept_
  m=-a1/a2
  c=-b/a2
  t=np.linspace(-4,6,100)
  is_correct = clf.predict(X_test) == y_test
  pred_class += is_correct

  acc=np.round(clf.score(X_test, y_test),2)
  #ax.scatter(t,m*t+c,s=10,alpha=0.2,marker=".",label="separation Fold {} - acc {}".format(i,acc))


pred_class/=n_fold
#pred_class = pred_class+1
im = ax.scatter(X_test[:,0],X_test[:,1],s=10,alpha=1,c=pred_class,label="Probability of beeing correctly classified ",cmap="seismic")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
f.colorbar(im, orientation="horizontal", pad=0.2)
```
